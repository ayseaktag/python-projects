```python
import random
picture = ["""
   +---+
   |   |
       |
       |
       |
       |
--------""","""
   +---+
   |   |
   O   |
       |
       |
       |
--------""","""
   +---+
   |   |
   O   |
   |   |
       |
       |
--------""","""
   +---+
   |   |
   O   |
  /|   |
       |
       |
--------""","""
   +---+
   |   |
   O   |
  /|\  |
       |
       |
--------""","""
   +---+
   |   |
   O   |
  /|\  |
  /    |
       |
--------""","""
   +---+
   |   |
   O   |
  /|\  |
  / \  |
       |
--------"""]
while True:
    print(("----->") + "Adam Asmaca Oyunu" + ("<-----"))
    
    word = random.choice(["python", "C", "angular", "react", "java", "html"])
    count = 0
    guess = []
   
    
    while True:
        print(picture[count])
        
        for i, char in enumerate(word):
            print(char if i in guess else "_"),
            
        answer = input("Kelimeyi Tahmin Edin: ")
        
        if answer == word:
            print("Kazandınız!")
            break
        else:
            while True:
                randomWord = random.randint(0, len(word))
                if not randomWord in guess:
                    guess.append(randomWord)
                    break
            count += 1
        
        if count >= len(picture):
            print("Kaybettiniz!")
            break
        
    if not "y" == input("Tekrar Oynamak İstermisiniz? (y/n): "):
        break
```
