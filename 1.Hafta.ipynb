Globalde tanımlanan price değerinin değiştirilmesi

```python
price = 500 #Global Değişken
def getSalary1():
  global price
  price = 600 #Yerel değişkenden global değişkene
  print(price)

def getSalary2():
  print(price)

print("1",price)
getSalary1()
print("2", price)
getSalary2()
print("3", price)
```

**While Döngüsü**

Belirli bir koşul doğru olduğunda while döngüsü çalışır. Koşul yanlışsa while döngüsünden çıkar, sıradaki döngüden devam eder.Koşul doğru olduğu sürece işlemlerin tekrarlanmasını sağlar.

```python
count = 1
while count < 6:
  print(count)
  count += 1 #count = count + 1
```

```python
number = int(input("10 ile 20 arasında bir sayı giriniz: "))
while number > 20 or number < 10: #or mantıksal operatördür. Ya da anlamına gelmektedir.
 print("İstenen değer aralığında bir sayı girmediniz!")
 number = int(input("Lütfen 10-20 aralığında bir değer giriniz. "))
else:
  print("Doğru aralıkta girilen değerimiz: ", number)
```
